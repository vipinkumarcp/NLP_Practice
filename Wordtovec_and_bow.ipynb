```python
#data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#stemming and stopwords
corpus = []
for i in range(0,len(messages)):
  # print(messages['message'][i])
  review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  # print(review)

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  # print(review)
  review = ' '.join(review)
  # print(review)
  corpus.append(review)



```

```python
len(corpus)
```

```python
#creating the bag of the words

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, binary=True,ngram_range=(2,2))
#It creates a binary bag-of-words representation with a limited vocabulary of the most frequent 2500 terms.
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y = pd.get_dummies(messages['lablel'])
```

```python
y
```

```python
y = y.iloc[:,1].values
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
#
```

```python
#prediction
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#stemming and stopwords
corpus = []
for i in range(0,len(messages)):
  # print(messages['message'][i])
  review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  # print(review)

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  # print(review)
  review = ' '.join(review)
  # print(review)
  corpus.append(review)
```

```python
from nltk import sent_tokenize
```

```python
from gensim.utils import simple_preprocess
```

```python
corpus[0]
```

```python
nltk.download('punkt_tab')
#simple process concerst a list of document into lowercase token
words = []
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
#This trains a Word2Vec model that learns word meanings using a context window of 5 and ignores rare words appearing less than twice.
model = gensim.models.Word2Vec(words,min_count=2,window=5)
```

```python
#It returns the vocabulary (list of learned words) from the Word2Vec model, ordered by frequency.
model.wv.index_to_key
```

```python
import numpy as np
#This function converts a document into a single vector by averaging the Word2Vec embeddings of its words.

def avg_word2vec(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.key_to_index]

    if len(vectors) == 0:
        return np.zeros(model.vector_size)

    return np.mean(vectors, axis=0)
```
