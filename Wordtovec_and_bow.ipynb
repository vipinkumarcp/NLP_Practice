```python
import pandas as pd
```

```python
messages =  pd.read_csv('/content/SMSSpamCollection',sep='\t',names=['lablel','message'])
```

```python
messages
```

```python
#data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#stemming and stopwords
corpus = []
for i in range(0,len(messages)):
  # print(messages['message'][i])
  review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  # print(review)

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  # print(review)
  review = ' '.join(review)
  # print(review)
  corpus.append(review)



```

```python
len(corpus)
```

```python
#creating the bag of the words

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, binary=True,ngram_range=(2,2))
#It creates a binary bag-of-words representation with a limited vocabulary of the most frequent 2500 terms.
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y = pd.get_dummies(messages['lablel'])
```

```python
y
```

```python
y = y.iloc[:,1].values
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
#
```

```python
#prediction
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#creating TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
#
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
