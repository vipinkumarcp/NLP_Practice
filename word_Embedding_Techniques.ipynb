```python
import tensorflow as tf
print(tf.__version__)
```

```python
from  tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
### Vocabulary size
voc_size=500
```

```python
#one hot encoding
#It loops through each word in sent and converts it into a numeric index using one_hot, creating a list of those indices.
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

```python
#word embedding represntation

from tensorflow.keras.layers import Embedding,Input
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential

```

```python
import numpy as np
```

```python
#to make all snetnce one hotting same size ..sentance leghn we can max legth of biggest sentence...pre padding zero in front
sent_length = 8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#now we need convert each and every value to convert vectors ...
#here i'm taking my feature representenation dimensions to 10
dim = 10
```

```python
#embedding layer similar to word to vec
model = Sequential()
model.add(Input(shape=(sent_length,)))
model.add(Embedding(voc_size,dim))
model.compile('adam','mse')
```

```python
model.summary
```

```python
model.predict(np.array([embedded_docs[1]]))
```

```python
model.predict(embedded_docs)
```
